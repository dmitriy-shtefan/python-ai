# Encoder. Summarization

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model_name = 'google/flan-t5-small'
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
article = """
Transformers changed NLP because they made it practical to train large models
on huge text corpora and capture long-range dependencies through attention.
"""
```

```python
prompt = "Summarize: " + article
```

```python
inputs = tokenizer(prompt, return_tensors="pt")
```

```python
outputs = model.generate(**inputs, max_new_tokens=40, do_sample=False)
```

```python
summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
summary
```

# Decoder. Transformer. GPT-2

```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="gpt2")
```

```python
outputs = generator(
    "Artificial intelligence can help small business by",
    max_new_tokens=50,
    do_sample=True,
    temperature=0.7,
    top_p=0.9
)
```
