```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1, 11)
y = 5 * x + 10
```

# Поняття

* Figure - контейнер, на якому будемо розташовувати графіки
* Axes - полотно, на якому будемо малювати графік(и)
* Axis - вісь, x, y

```python
plt.style.use('default')
plt.figure(figsize=(10, 6))

plt.plot(x, y, color='darkgreen', linestyle='-', linewidth=4, label='y = 5 * x + 10', marker='o', markersize=10)
plt.plot(x, y + 20, color='darkblue', linestyle='--', linewidth=4, label='y = 5 * x + 30', marker='s', markersize=10)
plt.plot(x, y + 40, color='red', linestyle='-.', linewidth=4, label='y = 5 * x + 50', marker='+', markersize=10)
plt.plot(x, y + 60, color='orange', linestyle=':', linewidth=4, label='y = 5 * x + 70', marker='^', markersize=10)

plt.title('Lines', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(alpha=0.3, axis='x')
plt.legend()
plt.show()
```

# Markers
* o - коло
* s - квадрат
* ^ - трикутник
* `+` - плюс

# Linestyles

* `-` - суцільна лінія
* `--` - преривчаста лінія
* `:` - лінія з крапок
* `-.` - преривчаста з крапкою

```python
x1 = np.linspace(1, 10, 1000)
y1 = np.sin(x1)

fig, axes = plt.subplots(figsize=(10, 6))

axes.plot(x1, y1, label='y = sin(x)')
axes.set_title('Demo Line')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.legend()
axes.grid(True)
```
