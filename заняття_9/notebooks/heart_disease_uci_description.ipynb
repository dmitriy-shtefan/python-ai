## Що означають колонки

| Колонка | Опис |
|---|---|
| `age` | вік пацієнта у роках |
| `sex` | стать: `1` - чоловік, `0` - жінка |
| `cp` | тип болю в грудях |
| `trestbps` | артеріальний тиск у стані спокою, мм рт. ст. |
| `chol` | рівень холестерину, мг/дл |
| `fbs` | рівень цукру натще понад 120 мг/дл: `1` - так, `0` - ні |
| `restecg` | результат електрокардіограми у стані спокою |
| `thalach` | максимальна досягнута частота серцевих скорочень |
| `exang` | стенокардія, спричинена фізичним навантаженням: `1` - так, `0` - ні |
| `oldpeak` | депресія ST після навантаження порівняно зі станом спокою |
| `slope` | нахил пікового сегмента ST під час навантаження |
| `ca` | кількість великих судин, забарвлених флюороскопією |
| `thal` | результат тесту thalassemia |
| `num` | цільова колонка: `0` - відсутність хвороби серця, `1-4` - наявність хвороби різного ступеня |

У домашньому завданні зручніше перетворити `num` на бінарну ціль: `0` - низький ризик, `1` - високий ризик.

````python
print("Типи даних:")
print(df.dtypes)

print()
print("Кількість пропусків у кожній колонці:")
print(df.isna().sum())
````

## На що звернути увагу перед навчанням

У цьому CSV є кілька пропусків у колонках `ca` та `thal`. Перед передаванням даних у PyTorch їх потрібно обробити: наприклад, заповнити медіаною або видалити рядки з пропусками.

Для домашнього завдання краще не видаляти колонки, бо всі вони є потенційними ознаками для моделі.

````python
print("Розподіл початкової цілі num:")
print(df["num"].value_counts().sort_index())

risk = (df["num"] > 0).astype("int64")
print()
print("Бінарна ціль:")
print(risk.value_counts().sort_index().rename(index={0: "0 - низький ризик", 1: "1 - високий ризик"}))
````

## Стартова підготовка `X` та `y`

Нижче показано мінімальну підготовку даних: окремо створюємо ознаки `X`, бінарну ціль `y` і заповнюємо пропуски медіанами.

Далі потрібно зробити `train/test split`, масштабування через `StandardScaler`, перетворення в тензори PyTorch і навчання MLP.

````python
X = df.drop(columns=["num"]).copy()
y = (df["num"] > 0).astype("int64")

X = X.fillna(X.median(numeric_only=True))

print("X shape:", X.shape)
print("y shape:", y.shape)
print("Пропуски після заповнення:", int(X.isna().sum().sum()))
X.head()
````
