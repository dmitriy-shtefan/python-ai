```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image = np.array(Image.open("../../data/cat.webp"))

print(image.shape)

# Візуалізація
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
contrast_factor = 2
less_contrast = np.clip(image / contrast_factor, 0, 255).astype(np.uint8)
```

```python
# Візуалізація
plt.imshow(less_contrast)
plt.axis("off")
plt.show()
```
