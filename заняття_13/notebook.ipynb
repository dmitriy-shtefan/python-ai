```python
import re
```

```python
text = "Great product!!! Delivery was late, but quality is AMAZING :)"

text = text.lower()
text = re.sub(r"[^a-z\s]", " ", text)
text = re.sub(r"\s+", " ", text)

print(text)
```

```python
import nltk
from nltk.tokenize import word_tokenize


nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
tokens = word_tokenize(text)
print(tokens)
```

"The", "capital", "of", "France" => x1, x2, x3, x4 => y1 = "is"

"The", "capital", "of", "France", "is" => x1, x2, x3, x4, x5 => y2 = "Paris"
