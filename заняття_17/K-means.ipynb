```python
# plotting the progression of 3 K means clustering on the same data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


from sklearn.cluster import KMeans

from IPython.display import display, HTML
```

```python

df = pd.read_csv("PointsCoordinates.csv")
```

```python
df
```

```python
data = df.iloc[:, 1:3].values
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

centroids_paths = [[], [], []]

def animate(i):
    ax.clear()
    
    if i == 0:
        ax.scatter(data[:, 0], data[:, 1], color='black', s=5)
        ax.set_title('Initial State')
        return
        
    kmeans = KMeans(n_clusters=3, init='random', n_init=1, max_iter=i, random_state=42)
    kmeans.fit(data)
    
    labels = kmeans.labels_
    
    ax.scatter(data[labels == 0][:, 0], data[labels == 0][:, 1], s=5, color='green', label='Cluster 1')
    ax.scatter(data[labels == 1][:, 0], data[labels == 1][:, 1], s=5, color='red', label='Cluster 2')
    ax.scatter(data[labels == 2][:, 0], data[labels == 2][:, 1], s=5, color='blue', label='Cluster 3')
    
    centers = kmeans.cluster_centers_
    for j, center in enumerate(centers):
        centroids_paths[j].append(center)
        path = np.array(centroids_paths[j])
        ax.plot(path[:, 0], path[:, 1], linestyle='--', color='black', linewidth=1)
        ax.scatter(center[0], center[1], color='black', s=100, marker='X')
    
    ax.set_title(f"Iteration {i}")
    ax.legend()


ani = animation.FuncAnimation(fig, animate, frames=11, repeat=False, interval=1000)

plt.close()
display(HTML(ani.to_jshtml()))
```
